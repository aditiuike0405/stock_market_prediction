```python

df['date'] = pd.to_datetime(df['date'])

df['Year'] = df['date'].dt.year
yearly_summary = df.groupby('Year').agg({'open': 'first', 'close': 'last', 'high': 'max', 'low': 'min'})
yearly_summary = yearly_summary.rename(columns={'open': 'Open Price', 'close': 'Close Price', 'high': 'High Price', 'low': 'Low Price'})
print("Year-wise opening,closing,and high price low price \n\n")
print(yearly_summary)
plt.figure(figsize=(12, 6))
yearly_summary.plot(kind='bar', colormap='viridis')
plt.title('Summary of Opening, Closing, High, and Low Prices by Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.show()
```

```python
sns.pairplot(df[['open', 'high', 'low', 'close', 'volume']], diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Variables')
plt.show()
```

```python
max_opening_price=df['open'].max()
min_opening_price=df['open'].min()
highest_open = df[df['open'] == max_opening_price]
lowest_open = df[df['open'] == min_opening_price]
plt.bar(highest_open['Name'], max_opening_price, label='Highest Open Price',color='yellow')
plt.bar(lowest_open['Name'], min_opening_price, label='Lowest Open Price',color='magenta')

plt.xlabel('Company')
plt.ylabel('Opening Price')
plt.title('Companies with Highest and Lowest Opening Prices')
plt.legend()

plt.show()

```

```python
max_closing_price=df['close'].max()
min_closing_price=df['close'].min()
highest_close = df[df['close'] == max_closing_price]
lowest_close = df[df['close'] == min_closing_price]
plt.bar(highest_close['Name'], max_opening_price, label='Highest Closing Price', color='green')
plt.bar(lowest_close['Name'], min_opening_price, label='Lowest Closing Price', color='pink')

plt.xlabel('Company')
plt.ylabel('Closing Price')
plt.title('Companies with Highest and Lowest Closing Prices')
plt.legend()

plt.show()

```

```python
sns.histplot(df['open'],bins=20, color='yellow')
plt.title('Histogram of Opening Prices with frequency')
plt.xlabel('Opening Price')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['close'],bins=20, color='pink')
plt.title('Histogram of Closing Prices with frequency')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Year', y='close', data=df, palette='muted')
plt.title('Violin Plot of Closing Prices by Year')
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['volume'], bins=30, kde=True, color='skyblue')
plt.title('Histogram of Daily Trading Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```

```python

top5_opening = df.groupby('Name')['open'].max().nlargest(5)
print("Top 5 Companies with Highest Opening Price:")
print(top5_opening)


top5_closing = df.groupby('Name')['close'].max().nlargest(5)
print("\nTop 5 Companies with Highest Closing Price:")
print(top5_closing)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

ax1.bar(top5_opening.index, top5_opening.values, color='skyblue')
ax1.set_title('Top 5 Companies with Highest Opening Price')
ax1.set_ylabel('Opening Price')
ax1.tick_params(axis='x', rotation=45)


ax2.bar(top5_closing.index, top5_closing.values, color='lightcoral')
ax2.set_title('Top 5 Companies with Highest Closing Price')
ax2.set_ylabel('Closing Price')
ax2.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python

top5_volume = df.groupby('Name')['volume'].sum().nlargest(5)


annual_volume_by_company = df.groupby(['Year', 'Name'])['volume'].sum().unstack().fillna(0)


plt.figure(figsize=(12, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen', 'lightyellow', 'lightblue']
annual_volume_by_company.plot(kind='bar', stacked=True, color=colors)
plt.title('Annual Trading Volume by Company')
plt.xlabel('Year')
plt.ylabel('Trading Volume')
plt.xticks(rotation=45)
plt.legend(title='Company', labels=top5_volume.index, loc='upper left')
plt.show()
```
